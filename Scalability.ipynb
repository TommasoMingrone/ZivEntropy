```python
import math
import random
import numpy as np
import matplotlib.pyplot as plt
from collections import Counter

# Seed for reproducibility
random.seed(0)

Cat1 = (2, 1, 1, 1)   # Chaotic
Cat2 = (1, 1, -1, 0)  # Regular

# ---------------------------------------------------------
# Unit square [0,1]x[0,1]
# ---------------------------------------------------------
def check_point(x, y):
   
    if 0 <= x <= 0.5 and 0 <= y <= 0.5:
        return 1
    if 0.5 <= x <= 1 and 0.5 <= y <= 1:
        return 1
    return 0

# ---------------------------------------------------------
# Arnold's Cat Map
# ---------------------------------------------------------
def arnold_cat_map_DATA(N, Cat):
    
    string = np.zeros(N, dtype=int)
    a, b, c, d = Cat
    x0, y0 = 0.1, 0.1  # punto di partenza

    for i in range(N):
        x = (a * x0 + b * y0) % 1
        y = (c * x0 + d * y0) % 1
        string[i] = check_point(x, y)
        x0, y0 = x, y

    return string

# ---------------------------------------------------------
# Computations and plotting
# ---------------------------------------------------------
def get_substring_frequencies(binary_list, d):

    counter = Counter()
    N = len(binary_list)
    total_windows = max(0, N - d + 1)
    
    for i in range(total_windows):
        pattern = tuple(binary_list[i : i + d])
        counter[pattern] += 1
    
    return counter, total_windows

def compute_entropy(counter, total_windows):
   
    if total_windows == 0:
        return 0.0
    H = 0.0
    for _, count in counter.items():
        p = count / total_windows
        H -= p * math.log2(p)
    return H

def compute_entropies_for_dims(binary_list, min_dim, max_dim):

    dims = []
    entropies = []
    N = len(binary_list)
    
    for d in range(min_dim, max_dim + 1):
        counter, total_windows = get_substring_frequencies(binary_list, d)
        H = compute_entropy(counter, total_windows)
        dims.append(d)
        entropies.append(H)
    
    return dims, entropies

def analyze_and_plot(binary_list, min_dim, max_dim, name="Sequence"):
  
    dims, entropies = compute_entropies_for_dims(binary_list, min_dim, max_dim)
    N = len(binary_list)
    
    print(f"\n{name} (length={N})")
    for d, Hval in zip(dims, entropies):
        print(f"  d={d}, H(d)={Hval:.4f}")
    
    alpha, beta = np.polyfit(dims, entropies, 1)
    print(f"  H(d) ~ {alpha:.4f} * d + {beta:.4f}")
    print(f"  => 'Lyapunov coefficient' ~ {alpha:.4f}")
    
    plt.figure(figsize=(6,4))
    plt.plot(dims, entropies, 'o-', label="H(d) data")
    plt.title(f"H(d) vs. d  {name} (N={N}, min_dim={min_dim}, max_dim={max_dim})")
    plt.xlabel("Sottostringa d")
    plt.ylabel("Entropia H(d)")
    
    d_line = np.linspace(min_dim, max_dim, 100)
    H_line = alpha * d_line + beta
    plt.plot(d_line, H_line, 'r--', label=f"Fit slope={alpha:.3f}")
    
    plt.legend()
    plt.tight_layout()
    plt.show()

# ---------------------------------------------------------
# List of parameters
# ---------------------------------------------------------
param_sets = [
    (1000,  2, 10),
    (20000,  2, 20),
    (50000,  2, 50),
    (100000,  2, 100),
    (500000,  2, 200)
    #(1000000,  2, 500) # Too slow
]

for i, (N, min_dim, max_dim) in enumerate(param_sets, start=1):
    print(f"\n=== Test {i} ===")
    print(f"N={N}, min_dim={min_dim}, max_dim={max_dim}")
    
    S_caotic = arnold_cat_map_DATA(N, Cat1)
    analyze_and_plot(S_caotic, min_dim, max_dim, name="Cat Chaotic")

    S_regular = arnold_cat_map_DATA(N, Cat2)
    analyze_and_plot(S_regular, min_dim, max_dim, name="Cat Regular")
```
