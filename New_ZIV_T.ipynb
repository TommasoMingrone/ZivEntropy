```python
import math
import random
import numpy as np
import matplotlib.pyplot as plt
from collections import Counter

random.seed(0)

N = 100000 #data length 
#min and max dimension of the substring
min_dim = 2
max_dim = 100
#k=10 #number of considered patterns per substring lenght

Cat1=(2,1,1,1) # Chaotic ; Lyapunov: (3 + sqrt(5)) / 2, (2.618) || (logarithm of the max eigenvalue of the matrix);
Cat2=(1,1,-1,0)# Regular

#GENERATE DATA

def check_point(x, y):
    """
    Divide the unit square [0,1]x[0,1] into four quadrants:
     - top-right and bottom-left => returns 1
     - top-left and bottom-right => returns 0
    """
    if 0 <= x <= 0.5 and 0 <= y <= 0.5:
        return 1
    if 0.5 <= x <= 1 and 0.5 <= y <= 1:
        return 1
    return 0

def arnold_cat_map_DATA(N, Cat):
    """
    Generates a binary sequence of length N from an Arnold's cat map iteration
    on the unit square. Each point (x,y) is mapped, and whether the point is in
    one of two 'quadrants' returns 1 or 0.

    Args:
        N   (int): Length of sequence to generate
        Cat (list or tuple): [a, b, c, d], the matrix coefficients for the cat map

    Returns:
        np.ndarray: A 1D array of 0s and 1s of length N
    """
    # Initialize array of length N
    string = np.zeros(N, dtype=int)

    a, b, c, d = Cat[0], Cat[1], Cat[2], Cat[3]
    
    # Starting point
    x0, y0 = 0.1, 0.1

    for i in range(N):
        # Apply cat map and reduce modulo 1
        x = (a * x0 + b * y0) % 1
        y = (c * x0 + d * y0) % 1

        # Check which quadrant and assign 0 or 1
        string[i] = check_point(x, y)

        # Update (x0, y0) for next iteration
        x0, y0 = x, y

    return string


S1=arnold_cat_map_DATA(N,Cat1) #Chaotic
S2=arnold_cat_map_DATA(N,Cat2) #Regular


# ------------------------------------------------------------------
# 1) Generate binary sequences (lists of ints [0 or 1])
# ------------------------------------------------------------------
def generate_alternating_sequence(N):
    """
    Returns a list of length N of the form [0,1,0,1,...].
    """
    return [(i % 2) for i in range(N)]

def generate_random_binary_sequence(N, seed=None):
    """
    Returns a random binary list of length N (0 or 1).
    If seed is specified, we set random.seed(seed) for reproducibility.
    """
    if seed is not None:
        random.seed(seed)
    return [random.randint(0, 1) for _ in range(N)]


# (A) Alternating sequence: [0,1,0,1, ...]
#S1 = generate_alternating_sequence(N)
# (B) Random sequence: [0/1,...]
#S2 = generate_random_binary_sequence(N, seed=42)
```

```python
# ------------------------------------------------------------------
# 2) Counting and Entropy
# ------------------------------------------------------------------
def get_substring_frequencies(binary_list, d):
    
    """
    Counts all length-d substrings in 'binary_list'.
    Each substring is stored as a tuple of ints (0/1).
    
    Returns:
      counter: Counter({tuple_of_ints: count})
      total_windows = (N - d + 1)
    If d > len(binary_list), returns an empty counter, 0 windows.
    """
    counter = Counter()
    total_windows = N - d + 1
    
    for i in range(total_windows):
        pattern = tuple(binary_list[i : i + d])
        counter[pattern] += 1
    
    return counter, total_windows

def compute_entropy(counter, total_windows):
    """
    Given the full distribution of all length-d substrings:
      - counter({pattern: count})
      - total_windows = total # of windows
    Compute the Shannon entropy (base 2):
      H = - sum(p(pattern) * log2(p(pattern))).
    """
    H = 0.0
    for pattern, count in counter.items():
        p = count / total_windows
        H -= p * math.log2(p)
    return H

# ------------------------------------------------------------------
# 3) Compute entropies for each dimension
# ------------------------------------------------------------------

def compute_entropies_for_dims(binary_list,min_dim, max_dim):
    """
    For each dimension d in [min_dim..max_dim]:
      1) Count frequencies of all substrings of length d
      2) Compute the exact entropy from the full distribution
    Returns:
      dims      = [d1, d2, ...]
      entropies = [H(d1), H(d2), ...]
    """
    dims = []
    entropies = []
    
    for d in range(min_dim, max_dim + 1):
        
        counter, total_windows = get_substring_frequencies(binary_list, d)
        
        H = compute_entropy(counter, total_windows)
        dims.append(d)
        entropies.append(H)
    
    return dims, entropies

# ------------------------------------------------------------------
# 4) Plotting routine
# ------------------------------------------------------------------

def analyze_and_plot(binary_list, name="Sequence"):
    """
    1) Compute entropies for d in [min_dim..max_dim].
    2) Fit H(d) = alpha*d + beta (una regressione lineare).
    3) Plot d vs. H(d) con la retta di best-fit sovrapposta.
    """
    # 1) Calculate the entropies for each dimension d
    dims, entropies = compute_entropies_for_dims(binary_list, min_dim, max_dim)

    print(f"\n{name} (length={len(binary_list)})")
    for d, Hval in zip(dims, entropies):
        print(f"  d={d}, H(d)={Hval:.4f}")

    # 2) Fit H(d) ~ alpha*d + beta
    alpha, beta = np.polyfit(dims, entropies, 1)
    print(f"  H(d) ~ {alpha:.4f} * d + {beta:.4f}")
    print(f"  => 'Lyapunov coefficient' ~ {alpha:.4f}")
    
    # 3) Plot d vs. H(d)
    plt.figure(figsize=(6, 4))
    plt.plot(dims, entropies, 'o-', label="H(d) data")
    plt.title(f"H(d) vs. dimension (fitted) — {name}")
    plt.xlabel("Substring dimension d")
    plt.ylabel("H(d)")

    d_min, d_max = min(dims), max(dims)
    d_line = np.linspace(d_min, d_max, 100)
    H_line = alpha * d_line + beta

    plt.plot(d_line, H_line, 'r--', label=f"Fit slope={alpha:.3f}")

    plt.legend()
    plt.tight_layout()
    plt.show()

    
analyze_and_plot(S1, name="Cat Chaotic")
analyze_and_plot(S2, name="Cat Regular")
```

```python
def pomeau_manneville_map_data(N, z=2.0, x0=0.1, threshold=0.5, burn_in=1000):
    """
    Generate a binary sequence of length N from the Pomeau–Manneville map:
      x_{n+1} = x_n + (x_n)^z mod 1.
    We skip an initial 'burn_in' steps to get rid of transients.
    
    Args:
      N         (int) : Desired length of the output binary sequence
      z         (float): exponent (default 2.0 for a standard case)
      x0        (float): initial condition
      threshold (float): for binarizing x in [0,1)
      burn_in   (int) : number of steps to discard for transients

    Returns:
      A list (or np.array) of length N with 0/1.
    """
    seq = []
    x = x0
    
    # primo burn-in per stabilizzare la dinamica
    for _ in range(burn_in):
        x = (x + x**z) % 1.0
    
    # adesso costruiamo la sequenza
    for _ in range(N):
        x = (x + x**z) % 1.0
        # partizioniamo con soglia (0.5 di default)
        bit = 0 if x < threshold else 1
        seq.append(bit)
    
    return seq
```

```python
S_pomeau = pomeau_manneville_map_data(N, z=2.0, x0=0.1, threshold=0.5, burn_in=1000)

analyze_and_plot(S_pomeau, name="Pomeau–Manneville (z=2.0)")
```
